```python
# import pandas and csv
import pandas as pd
vine_table_df=pd.read_csv("Resources/vine_table.csv",delimiter=None)
vine_table_df.head()
```

```python
# examine data types
vine_table_df.dtypes
```

```python
# Filter the data total_votes count is equal to or greater than 20
great_vote_df=vine_table_df.loc[vine_table_df["total_votes"]>=20]
great_vote_df.head()
```

```python
# Filter the data from previous step where the number of helpful_votes divided by total_votes 
# is equal to or greater than 50%
rate_df=great_vote_df.loc[(great_vote_df["helpful_votes"]/great_vote_df["total_votes"])>=0.5]
rate_df.head()
```

```python
# Filter the data from previous step where a review was written as part of the Vine program (paid), vine == 'Y'
paid_df=great_vote_df.loc[great_vote_df["vine"]=="Y"]
paid_df.head()
```

```python
# Filter the data where a review was written as part of the Vine program (paid), vine == 'N'
unpaid_df=great_vote_df.loc[great_vote_df["vine"]=="N"]
unpaid_df.head()
```

```python
def creatsummary (df):
    review_number=df["review_id"].count()
    five_star=df["review_id"].loc[df["star_rating"]==5].count()
    five_star_percetage=five_star/review_number
    summary=pd.DataFrame([{"Review_number":review_number,"Five_star":five_star,"Five_star_percetage":five_star_percetage}],index=["Result"])
    summary["Review_number"]=summary["Review_number"].map("{:,.0f}".format)
    summary["Five_star"]=summary["Five_star"].map("{:,.0f}".format)
    summary["Five_star_percetage"]=(summary["Five_star_percetage"]*100).map("{:,.1f}%".format)
    return(summary)
```

```python
# Paid summary of total number of reviews, number of 5-star reviews, percetage of 5-star reviews
creatsummary (paid_df)
```

```python
# Unpaid summary of total number of reviews, number of 5-star reviews, percetage of 5-star reviews
creatsummary (unpaid_df)
```
